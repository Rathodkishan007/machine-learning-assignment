```python
#Initialising the libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, classification_report, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
#Loading the values of telecom churn data
telecom_data=pd.read_csv("Telecom_Churn_Data.csv")
```

```python
#Column datatype for telecom data
telecom_data.info()
```

```python
#Null values has to be replaced. We would be using a 0 to replace that
telecom_data=telecom_data.fillna(0)
```

```python
#Independent variables dataset and dependent variable dataset creation
indep_vrbl = telecom_data.drop(['Churn_Value'],axis=1)
dep_vrbl = telecom_data.loc[:,'Churn_Value']
```

```python
#Scaling the independent variable dataset
norm= MinMaxScaler((-1,1))
indep_vrbl_norm=norm.fit_transform(indep_vrbl)
```

```python
#Training dataset and test dataset creation
X_trn,X_tst,Y_trn,Y_tst = train_test_split(indep_vrbl_norm, dep_vrbl ,test_size=0.25,random_state=10)
```

```python
#Machine Learning 1st Technique- K Nearest Neighbour
import warnings
with warnings.catch_warnings(record=True):
    knn_ml = KNeighborsClassifier(n_neighbors = 4, metric = 'minkowski', p = 3)
    knn_ml.fit(X_trn, Y_trn)
    predct_op1 = knn_ml.predict(X_tst)
    print('R-squared score for 1st Technique- K Nearest Neighbour -->',r2_score(Y_tst, predct_op1))
    print('Mean Absolute Error for 1st Technique- K Nearest Neighbour  -->', mean_absolute_error(Y_tst, predct_op1))
    print('Root Mean Squared Error for 1st Technique- K Nearest Neighbour  -->', np.sqrt(mean_squared_error(Y_tst, predct_op1)))
    print('Classification Report for 1st Technique- K Nearest Neighbour  -->\n', classification_report(Y_tst, predct_op1))
```

```python
#Machine Learning 2nd Technique- Logistic Regression
import warnings
with warnings.catch_warnings(record=True):
    lr_ml =LogisticRegression(random_state=5)
    lr_ml.fit(X_trn, Y_trn)
    predct_op2 = lr_ml.predict(X_tst)
    print('R-squared score for 2nd Technique- Logistic Regression -->',r2_score(Y_tst, predct_op2))
    print('Mean Absolute Error for 2nd Technique- Logistic Regression  -->', mean_absolute_error(Y_tst, predct_op2))
    print('Root Mean Squared Error for 2nd Technique- Logistic Regression  -->', np.sqrt(mean_squared_error(Y_tst, predct_op2)))
    print('Classification Report for 2nd Technique- Logistic Regression  -->\n', classification_report(Y_tst, predct_op2))
```

```python
#Machine Learning 3rd Technique- Gaussian Naive Bayes
import warnings
with warnings.catch_warnings(record=True):
    gnb_ml =GaussianNB()
    gnb_ml.fit(X_trn, Y_trn)
    predct_op3 = gnb_ml.predict(X_tst)
    print('R-squared score for 3rd Technique- Gaussian Naive Bayes -->',r2_score(Y_tst, predct_op3))
    print('Mean Absolute Error for 3rd Technique- Gaussian Naive Bayes  -->', mean_absolute_error(Y_tst, predct_op3))
    print('Root Mean Squared Error for 3rd Technique- Gaussian Naive Bayes  -->', np.sqrt(mean_squared_error(Y_tst, predct_op3)))
    print('Classification Report for 3rd Technique- Gaussian Naive Bayes -->\n', classification_report(Y_tst, predct_op3))
```

```python
#Machine Learning 4th Technique- PCA Clustering
pca = PCA()
vls = indep_vrbl.values
pca.fit(vls)
churn_pca = pca.transform(vls)
```

```python
#Plot for PC1 and PC2
axis = plt.gca()
axis.scatter(churn_pca[:,0], churn_pca[:,1])
axis.set_xlabel('PC 1')
axis.set_ylabel('PC 2')
```

```python
#Plot for finding optimal cluster count
import warnings
with warnings.catch_warnings(record=True):
    axis = plt.gca()
    explain_var = pca.explained_variance_ratio_
    axis.plot(explain_var, marker='x')
    axis.set_xlabel('PC')
    axis.set_ylim(0,1.)
    axis.set_xticklabels(["PC{}".format(i) for i in range(8)])
```

```python
#Variance explained by PC's
print("Variance explained by PC's->",explain_var)
```
